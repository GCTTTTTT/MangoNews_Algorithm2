```python
# single-pass-ByTitle-STranformer-LOAD   0.99
import matplotlib.pyplot as plt
import pandas as pd

# 从文本文件中解析数据
data = {
    'threshold': [],
    'RI': [],
    'Precision': [],
    'Recall': [],
    'F-value': [],
    'ARI': [],
}

with open('./EVAL-single-pass-ByTitle-STranformer-LOAD_results.txt', 'r') as f:
    for line in f:
        line = line.strip()
        if line.startswith('threshold:'):
            data['threshold'].append(float(line.split()[-1]))
        elif line.startswith(('RI', 'Precision', 'Recall', 'F-value', 'ARI')):
            key, value = line.split(': ')
            data[key].append(float(value))

# 将数据转换为pandas DataFrame
df = pd.DataFrame(data)

# 根据threshold列进行排序
df = df.sort_values(by='threshold')
print(df)

# 绘制图表
plt.figure(figsize=(10, 6))  # 设置图表大小
for column in df.columns:
    if column != 'threshold':
        plt.plot(df['threshold'], df[column], marker='o', label=column)  # 添加点标记

plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Metric Scores by Threshold')  # 添加图表标题
plt.legend()
plt.grid(True)  # 添加网格线
plt.tight_layout()  # 自动调整子图参数, 使之填充整个图像区域
plt.show()
```
