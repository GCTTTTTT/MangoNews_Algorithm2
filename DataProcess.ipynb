```python
# 统计不同日期及对应的数据数量
import pandas as pd

file_name = './datasets/deduplicated_mangoNews_Nums3000p_CategoryMerge_new_undersampled.csv'

# 读取CSV文件
df = pd.read_csv(file_name,low_memory=False,lineterminator="\n")

# 将pub_time列转换为日期时间格式
df['pub_time'] = pd.to_datetime(df['pub_time'])

# 按照日期进行分组并统计每个日期分组中的记录数量
date_counts = df.groupby(df['pub_time'].dt.date).size()

# 将统计数据转换为DataFrame对象
date_counts_df = date_counts.reset_index(name='count')

# 按日期越新的顺序排序
date_counts_df = date_counts_df.sort_values(by='count', ascending=False)


# 保存统计数据到CSV文件
date_counts_df.to_csv('Data_counts_new_sorted.csv', index=False)

# pd.set_option('display.max_rows', None)
# 打印结果
print(date_counts_df)
```

```python
# 提取最新的10天的新闻：2023-12-02-2023-12-11
import pandas as pd


file_name = './datasets/deduplicated_mangoNews_Nums3000p_CategoryMerge_new_undersampled.csv'

# 读取CSV文件
df = pd.read_csv(file_name,low_memory=False,lineterminator="\n")

# 将pub_time列转换为日期时间格式
df['pub_time'] = pd.to_datetime(df['pub_time'])

# 筛选日期在2023-12-02至2023-12-11之间的数据
filtered_df = df[(df['pub_time'] >= '2023-12-02') & (df['pub_time'] <= '2023-12-11')]

# 按照日期进行排序
sorted_df = filtered_df.sort_values(by='pub_time')

# 将结果保存到新的CSV文件中
sorted_df.to_csv('Data231202-231211.csv', index=False)

# 打印结果
print("已将日期在2023-12-02至2023-12-11之间的数据保存到 filtered_data.csv 文件中。")
sorted_df
```
